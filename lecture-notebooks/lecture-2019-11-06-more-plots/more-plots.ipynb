```python
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
#
# overlay line plot with axes with data points
#

# sin curve
x1 = np.linspace(-5, 5, 100)
y1 = np.sin(x1)

# horizontal axis
x2 = np.array([-5, 5])
y2 = np.array([0, 0])

# vertical axis
x3 = np.array([2, 2])
y3 = np.array([-1, 1])

# data point 1
x4 = np.pi
y4 = 0

# data point 2
x5 = 2.0
y5 = np.sin(x5)
```

```python
pyplot.plot(x1, y1);
pyplot.plot(x2, y2, linestyle='--', color='Gray');
pyplot.plot(x3, y3, linestyle='--', color='Gray');
pyplot.scatter([x4, x5], [y4, y5], s=100);
pyplot.text(x4 + 0.5, y4 + 0.1, "datapoint 4");
pyplot.text(x5 + 0.5, y5, "datapoint 5");
pyplot.arrow(x4, y4, (x5-x4), (y5-y4), 
             head_width=0.3, 
             head_length=.2, 
             length_includes_head=True,
             color='red');
```

# Fill between

```python
x = np.linspace(-10, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
pyplot.plot(x, y_sin)
pyplot.plot(x, y_cos);
```

```python
pyplot.plot(x, y_sin)
pyplot.plot(x, y_cos);
pyplot.fill_between(x, y_sin, y_cos, where=y_sin < y_cos);
```

# Bar plot

```python
x = np.linspace(-2, 5, 10)
y1 = np.random.rand(10)
y2 = y1 + np.random.rand(10)
```

```python
width = 0.25
pyplot.bar(x, y1, width=width)
pyplot.bar(x+width, y2, width=width)
pyplot.xlim((-5, 5))
```

```python
width = 0.25 * 2
pyplot.bar(x, y1, width=width)
pyplot.bar(x, y2, width=width, bottom=y1)
```
