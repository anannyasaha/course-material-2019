# Plot some lines

```python
jack = [10, 20, 15, 14, 12, 10]
jill = [12, 18, 16, 17, 10, 13]
```

```python
pl.plot(jack, linestyle=':', marker='o', label="Jack's grades")
pl.plot(jill, linestyle='-.', marker='o', label="Jill's grades")
pl.legend()
pl.title('Grades over time')
pl.xlabel('Quiz')
pl.ylabel('Final mark');
```

# Scatter plot

```python
import numpy as np
```

```python
xs = [np.random.randn() for i in range(1000)]
ys = [np.random.randn() for i in range(1000)]
sz = [np.random.random() * 50 for i in range(1000)]
```

```python
pl.scatter(xs, ys, s=sz, alpha=0.4);
pl.grid(True);
```

```python
pl.scatter(xs, ys, s=50, alpha=0.3, c=sz, cmap=pl.get_cmap('jet'), edgecolors=None);
```

# Use numpy and matplotlib together

```python
import numpy as np
```

```python
course_grades = np.array([[1, 10, 8],
                         [2, 9, 12],
                         [3, 20, 18],
                         [4, 10, 21],
                         [5, 7, 12]], dtype=float)
```

```python
course_grades.shape
```

```python
pl.plot(course_grades[:, 0], course_grades[:, 1], label='Jack');
pl.plot(course_grades[:, 0], course_grades[:, 2], label='Jill');
pl.legend();
```

```python
center1 = np.array([-1, 1])
blob1 = np.random.randn(1000, 2) * 0.5 + center1
d1 = np.linalg.norm(blob1 - center1, axis=1)
```

```python
center2 = np.array([1, -1])
blob2 = np.random.randn(1000, 2) * 0.5 + center2
d2 = np.linalg.norm(blob2 - center2, axis=1)
```

```python
cm = pl.get_cmap('jet')
pl.scatter(blob1[:, 0], blob1[:, 1], c=d1, cmap=cm, alpha=0.4, edgecolors=None)
pl.scatter(blob2[:, 0], blob2[:, 1], c=d2, cmap=cm, alpha=0.4, edgecolors=None)
pl.grid(True);
```
